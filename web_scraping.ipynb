```python
import requests
from bs4 import BeautifulSoup
import string
```

```python
enter_search = input("votre recherche = ")
u_i = string.capwords(enter_search)
lists = u_i.split()
word ='_'.join(lists)
link = "https://en.wikipedia.org/wiki/"+word
print(link)
```

```python
def getdata(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup
print(getdata(link))
```

```python
def gettitle(soup):
    title = soup.title
    return title
print(gettitle(getdata(link)).string)
```

```python
def getpara(soupe):
    for para in soupe.find_all("p"):
        print(para.get_text())
    return para 
print(getpara(getdata(link)))
        
```

```python
def getlink(soupe):
    for link in soupe.find_all("a"):
        print(link.get("href"))
    return link
print(getlink(getdata(link)))
```

```python
def scraping(link):
    getdata(link)
    print("title is : ", gettitle(getdata(link)).string)
    print(getpara(getdata(link)))
    print(getlink(getdata(link)))

scraping(link)
```
